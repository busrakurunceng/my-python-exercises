# Birleştirme (Join)

```python
import numpy as np
import pandas as pd
m=np.random.randint(1,30, size=(5,3))
df1= pd.DataFrame(m ,columns = [ "var1" , "var2" , "var3"])
```

```python
df1
```

```python
df2=df1+99
df2
```

```python
pd.concat([df1,df2]) #indexler sıkıntılı
```

```python
pd.concat([df1,df2] , ignore_index=True) #ignore=görmezden gelmek
```

```python
df1.columns
```

```python
df2.columns=["var1" , "var2" , "deg3"]
```

```python
df2
```

```python
pd.concat([df1,df2] , ignore_index=True)
```

```python
pd.concat([df1,df2] , join="inner" ,ignore_index=True) 
# aynı değer türlerini birleştirmek için join kullanılır
```

```python
pd.concat([df1,df2] , join_axes = [df2.columns]) 
```

# İleri Birleştirme

```python
import pandas as pd
```

```python
#birebir birleştirme
```

```python
df1=pd.DataFrame({'calisanlar': ['Ali' ,'Veli' ,'Ayse' ,'Fatma' ] ,
                  'grup': ['Muhasebe','Muhendislik','Muhendislik','İk']})
df1
```

```python
df2=pd.DataFrame({'calisanlar': ['Ali' ,'Veli' ,'Ayse' ,'Fatma' ] ,
                  'ilk_giris': [2010,2009,2014,2019]})
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2 , on="calisanlar") # calisanlar üzerinde birleştirme yap diye belirtiliyor
```

```python
#coktan teke
df3=pd.merge(df1,df2)
df3
```

```python
df4=pd.DataFrame({ 'grup': ['Muhasebe'  , 'Muhendislik' , 'İk'] ,
                 'mudur' : ['Caner' , 'Mustafa' , 'Büşra']})
df4
```

```python
pd.merge(df3,df4 , on="grup")
```

```python
#coktan coka
df6=pd.DataFrame({'grup': ['Muhasebe' , 'Muhasebe' , 'Muhendislik' , 'Muhendislik' ,'İk', 'İk'] ,
                 'yetenekler': ["matematik" , "excel" , "kodlama" , "linux" , "excel" , "yonetim"]})
df6
```
