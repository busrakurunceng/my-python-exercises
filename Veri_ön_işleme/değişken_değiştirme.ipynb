# Veri standardizyonu

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,5,7])
V2 = np.array([7,7,5,8,12])
V3 = np.array([6,12,5,6,14])

df=pd.DataFrame( { "V1" :V1 , "V2" :V2 , "V3" :V3 })

df=df.astype(float)
df
```

## standardizyon

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df)
```

```python
df.mean()
```

## normalizasyon

```python
preprocessing.normalize(df)
```

## min max dönüşümü

```python
scaler = preprocessing.MinMaxScaler(feature_range= (10,20))
```

```python
scaler
```

```python
scaler.fit_transform(df)
```

# ONE-HOT Dönüşümü ve Dummy Değişken Tuzağı

```python
df.head()
```

```python
df_one_hot = pd.get_dummies(df , columns = ["sex"] , prefix= ["sex"])
```

```python
df_one_hot.head()
```

"sex_Male" ve "sex_Female" sütunları incelenmeli.  "sex" değişkeni silinerek bu iki değişkene dönüştü. Artık satır bazında bakılınca 1 ise cinsiyet odur. mesela 0.indekste kadın. 

Kategorik değişkenleri nümerik değişkenlere  dönüştürmek istediğimiz zaman "One-Hot" dönüşümünü kullanacağız. Fakat burada da dummy değişken tuzağı problemi söz konusu. 

"dummy değişken tuzağı" : Eğer veri seti içerisindeki değişkenlere uygulanan bir dönüşüm sonrasında  oluşturulan yeni değişkenler birbirleri üzerinden oluşturulabiliyorsa denir.

Yani bir değişkeni ifade eden başka bir değişken varsa "dummy değişken tuzağı" var.

Mesela male ve female değişkeni birbirlerini ifade edebilir.

__Kategorik değişkenin sınıf sayısından daha az sayıda dummy değişkeni olmalı ve değişkenin asıl hali de veri setinde bulunmamalı__

```python
df_one_hot_day = pd.get_dummies(df , columns = ["day"] , prefix= ["day"])
```
